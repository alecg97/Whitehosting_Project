```python
#pip install sentence-transformers==2.2.2
from sentence_transformers import SentenceTransformer, util
#pip install torch
import torch

model = SentenceTransformer('intfloat/multilingual-e5-base')
```

```python
def _calculate_embeddings(sentences):
    embeddings = model.encode(sentences, show_progress_bar=True, device=model.device)
    embeddings = torch.tensor(embeddings)
    embeddings = embeddings.to(model.device)
    embeddings = util.normalize_embeddings(embeddings)
    return embeddings.cpu().detach().numpy()
```

```python
def calculate_sentence_embedding(sentence):
    return _calculate_embeddings([sentence])[0]
```

```python
def question_and_answer(pregunta, productos_recomendados=5, verbose=False):
    """
    Retrieves recommended products based on a given question.

    Parameters:
    - pregunta (str): The question to be used for retrieving recommendations.
    - productos_recomendados (int): The number of recommended products to retrieve (default: 5).
    - verbose (bool): Whether to print additional information (default: False).

    Returns:
    - str: The response to the question.

    """
    emb_pregunta = calculate_sentence_embedding(pregunta)
    
    similitudes = []
    for e in embeddings:
        similitudes.append(cosine_similarity(e.reshape(1, -1), emb_pregunta.reshape(1, -1))[0][0])

    ind = np.argpartition(np.array(similitudes), -productos_recomendados)[-productos_recomendados:]

    sol = data.loc[ind]
    info_prods = ''
    i = 1
    for _, fila in sol.iterrows():
        info_prod = f'{i}.- Nombre: {fila["name"]} - DescripciÃ³n: {fila["description"]} - Nombre del host: {fila["host_name"]} - Barrio: {fila["neighbourhood"]} - Tipo de propiedad: {fila["property_type"]}\n' # AÃ±ade aquÃ­ las columnas quieres que se utilicen para poder preguntar
        info_prods += info_prod
        i += 1
    
    if verbose:
        print('\nPrompt utilizado:')
        print(sysprompt.format(info_prod=info_prods))
        print(pregunta)
    return preguntar_llama(sysprompt.format(info_prod=info_prods), pregunta)
```
